```python
from IPython.display import clear_output

class Game:
    
    def __init__(self):
        self.board = {
        '7': ' ',
        '8': ' ',
        '9': ' ',
        '4': ' ',
        '5': ' ',
        '6': ' ',
        '1': ' ',
        '2': ' ',
        '3': ' '
        }
        self.player1 = None
        self.player2 = None
        self.active_player = None
        self.player1_moves = set()
        self.player2_moves = set()
        self.winning_combinations = [
            {'7', '8', '9'},
            {'4', '5', '6'},
            {'1', '2', '3'},
            {'7', '4', '1'},
            {'8', '5', '2'},
            {'9', '6', '3'},
            {'7', '5', '3'},
            {'9', '5', '1'}
        ]
        self.all_moves ={}

    def initiate(self):
        print("Welcome to Tic Tac Toe")
        player1 = input("Player 1: Do you want to be X or O")
        player1 = player1.upper()
        print("Player 1 will go first.")
        self.player1 = player1
        self.player2 = 'O' if player1 == 'X' else 'X'
        self.play_game()


    

    def display_board(self):
        """
        Displays the current state of the board in a numpad layout.
        """
        print(f" {self.board['7']} | {self.board['8']} | {self.board['9']} ")
        print("-----------")
        print(f" {self.board['4']} | {self.board['5']} | {self.board['6']} ")
        print("-----------")
        print(f" {self.board['1']} | {self.board['2']} | {self.board['3']} ")

    

    def check_winners(self, player_moves):
        """
        Checks if the player's moves contain any winning combination.

        Args:
            player_moves (set): The set of positions the player has selected.

        Returns:
            bool: True if the player has won, False otherwise.
        """
        for combination in self.winning_combinations:
            if combination.issubset(player_moves):
                return True
        return False

    
    def valid_move(self, move):
        while True:
            if move in self.all_moves:
                move = input(f"{self.active_player} That position is already filled. Pick a different position other than {list(self.all_moves.keys())}, Draw in any position from 1-9: ")
            elif move not in self.board:
                move = input(f"{self.active_player} Pick position within range of 1-9: ")
            else:
                break
        return move
    
        

    def play_game(self):
        for turn in range(1,10):
            self.active_player = self.player1 if turn % 2 != 0 else self.player2
            player_moves = self.player1_moves if self.active_player == self.player1 else self.player2_moves
            self.display_board()

            move = input(f"Player {1 if turn % 2 != 0 else 2} Its your Turn, Draw {self.active_player} in any position from 1-9")
            move = self.valid_move(move)
            self.all_moves[move] = 1
            player_moves.add(move)
            self.board[move] = self.active_player
            
            clear_output(wait=True)
            print(f"Occupied positions:{list(self.all_moves.keys())}")
            self.display_board()

            if self.check_winners(player_moves):
                print(f"Congratulations! Player {1 if turn % 2 != 0 else 2} wins")
                break

            if turn == 9:
                print("Game Draws")

        retry = input("Do you want to restart the Game? Say yes or no:").lower()
        if retry =="yes":
            self.__init__()
            self.initiate()

game = Game()
game.initiate()





   
```
